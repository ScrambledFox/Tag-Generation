```python
import os
import os.path as path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

PATH = path.join('..', 'data', 'dataframe.csv')
df = pd.read_csv(PATH)

IMAGE_PATH = path.join('..', 'images')
os.makedirs(IMAGE_PATH, exist_ok=True)
```

```python
# Count how many rows have a value of 'True' of columns that start with 'TAG_'
# print(df.filter(regex='^TAG_').sum(axis=1).value_counts())
tags_df = df.filter(regex='^TAG_')

print(tags_df.sum(axis=0))
```

```python
# Plot the number of tags per row
tag_sum = tags_df.sum(axis=0)
tag_sum.sort_values(ascending=False, inplace=True)

plt.bar(tag_sum.index, tag_sum.values)
plt.xticks(rotation=90)
plt.ylabel('Number of rows')
plt.title('Number of tags per row')
plt.savefig(path.join(IMAGE_PATH, 'tag_sum.png'), bbox_inches='tight')
plt.show()
```
