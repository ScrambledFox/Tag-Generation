```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 200, 200)
y = x 
plt.plot(x, y, color='black', label='y=x')

plt.xlim(40, 160)
plt.ylim(40, 160)

plt.xlabel('Actual')
plt.ylabel('Predicted')

plt.annotate("MSE: {:.3f}".format(0), (0.7, 0.25), xycoords='axes fraction')
plt.annotate("RMSE: {:.3f}".format(0), (0.7, 0.20), xycoords='axes fraction')
plt.annotate("MAE: {:.3f}".format(0), (0.7, 0.15), xycoords='axes fraction')
plt.annotate("R2: {:.3f}".format(0), (0.7, 0.10), xycoords='axes fraction')
plt.annotate("MAPE: {:.2f}%".format(0 * 100), (0.7, 0.05), xycoords='axes fraction')

plt.savefig(os.path.join('..', 'images', 'empty_plot.png'))

plt.show()
```
